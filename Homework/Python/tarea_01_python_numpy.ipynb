**P8.** Crea dos arreglos de 5 elementos con `np.random.randint` (valores entre 1 y 20).  
Súmalos, réstalos y multiplícalos elemento por elemento. Imprime cada resultado.

```python
# Tu respuesta aquí
print("Vectores:")
array1 = np.random.randint(1,20,5)
print(array1)
array2 = np.random.randint(1,20,5)
print(array2)
array3 = np.random.randint(1,20,5)
print(array3)
print("----------------------")
print("Suma de vectores:")
print(array1+array2+array3)
print("----------------------")
print("Resta de vectores:")
print(array1-array2-array3)
print("----------------------")
print("Multiplicacion de vectores:")
print(array1*array2*array3)
print("----------------------")
```

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,  
usa **indexación booleana** para obtener solo las notas mayores o iguales a 6. Imprime el resultado.

```python
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

# Tu respuesta aquí
mayores_iguales = notas[notas >=6]
print("Notas Mayores o iguales a 6")
print(mayores_iguales)
```
